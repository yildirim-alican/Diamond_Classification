```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype
```

```
sns.set(rc={'figure.figsize':(13,10)})
```

```
df_diamonds=sns.load_dataset('diamonds')
df_diamonds.head()
```

```
df_diamonds.info()
```

```
df_diamonds.columns
```

```
df_diamonds["cut"].unique()
```

```
df_diamonds["cut"].value_counts()
```

```
df_diamonds["cut"].head()
```

```
df_diamonds["cut"].value_counts().plot.barh().set_title("Cut Variables")
```

```
df_diamonds["cut"].value_counts().plot.bar().set_title("Cut Variables 2")
plt.figure(figsize=(8,6))
```

```
sns.barplot(x="cut", y=df_diamonds["cut"].index, data=df_diamonds, color="red").set_title("Cut Density")
```

```
sns.barplot(x="cut", y="price", hue="color", data=df_diamonds)
plt.title("Cut  Price Color")
```

```
sns.barplot(x="cut",
            y="price",
            hue="clarity",
            data=df_diamonds)
plt.title("Cut  Price Clarity")
```

```
df_diamonds.info()
```

```
df_diamonds.groupby("cut").sum().plot.pie( y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(5,5))
```

```
cols = ["x", "y", "z"]

for col in cols:
  plt.figure(figsize=(20,5))
  sns.scatterplot(x=col, y="price", data=df_diamonds)
```

```
sns.histplot(df_diamonds["price"], kde=True)
```

```
sns.histplot(df_diamonds["price"], kde=True, bins=100)
```

```
sns.kdeplot(df_diamonds["price"])
```

```
sns.jointplot(x="price", y=df_diamonds["carat"], color="purple", data=df_diamonds)
```

```
sns.catplot(x="cut", y="price", color="red", data=df_diamonds)
plt.title("Cut - Price")
```

```
sns.catplot(x="cut", y="price", hue="color", kind="point", data=df_diamonds)
```

```
sns.kdeplot(df_diamonds["price"], shade=True, color="blue")
```

```
(sns.FacetGrid(df_diamonds, hue="cut", height=5, xlim=(0, 10000)).map(sns.kdeplot, "price", shade=True).add_legend())
```
